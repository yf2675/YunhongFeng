```python
import numpy as np
import pandas as pd
```

```python
# Loading Data
file = 'U.S._Chronic_Disease_Indicators__CDI_.csv'
df = pd.read_csv(file)   
df = pd.DataFrame(df)
df
```

```python
# Question 1: Filter out needed variables
usefulData1 = ["YearStart", "LocationAbbr", "LocationDesc", "Question", "Stratification1", "DataValue"]
usefulData2 = df["Stratification1"].isin(['Male','Female'])
usefulData = df[usefulData2][usefulData1]

usefulData3 = usefulData["Question"].isin(['Binge drinking prevalence among adults aged >= 18 years', 'Poverty'])
usefulData = usefulData[usefulData3]
usefulData
```

```python
# Question 2:  Change into wide format 
usefulData = pd.get_dummies(usefulData, columns=["Question"])
usefulData["binge_male"] = usefulData.loc[usefulData["Stratification1"] == "Male",["Question_Binge drinking prevalence among adults aged >= 18 years"] ]
usefulData["binge_female"] = usefulData.loc[usefulData["Stratification1"] == "Female",["Question_Binge drinking prevalence among adults aged >= 18 years"] ]

usefulData["binge_all"] = usefulData.loc[usefulData["Question_Binge drinking prevalence among adults aged >= 18 years"] == 1, ["DataValue"]]
usefulData["binge_male"] = usefulData.loc[usefulData["binge_male"] == 1, ["DataValue"]]
usefulData["binge_female"] = usefulData.loc[usefulData["binge_female"] == 1, ["DataValue"]]
usefulData["poverty"] = usefulData.loc[usefulData["Question_Poverty"] == 1, ["DataValue"]]

usefulData
```

```python
# Question 3: Rename varaibles and export data 
finalData = usefulData[["LocationDesc", "LocationAbbr", "YearStart", "binge_all", "binge_male", "binge_female", "poverty"]]
finalData = finalData.set_axis(["state", "stateabb", "year", "binge_all", "binge_male", "binge_female", "poverty"], axis=1)
finalData

finalData.to_csv("binge_clean.csv")
```

```python
# Question 4: Top10 Drinking states
finalData.sort_values("year", ascending = False)
finalData_2021 = finalData[finalData.year == 2021]

finalData_2021["binge_all"] = finalData_2021['binge_all'].astype(float)
finalData_2021["binge_male"] = finalData_2021['binge_male'].astype(float)
finalData_2021["binge_female"] = finalData_2021['binge_female'].astype(float)
finalData_2021["poverty"] = finalData_2021['poverty'].astype(float)

finalData_2021 = finalData_2021.groupby("state")[["binge_all", "binge_male", "binge_female"]].mean().reset_index()
finalData_2021_top10 = finalData_2021.sort_values("binge_all", ascending = False)
finalData_2021_top10 = finalData_2021_top10.iloc[:10]
finalData_2021_top10
```

```python
finalData["yeargap"] = finalData.loc[~finalData.binge_all.isna(), "year"]
finalData

growthrateData = finalData.sort_values("year", ascending = True).groupby("state")["yeargap","binge_all"].agg(["first", "last"]).reset_index()
growthrateData = growthrateData.set_axis(["state", "yearfirst", "yearlast", "bingefirst", "bingelast"], axis=1)

growthrateData["yearfirst"] = growthrateData["yearfirst"].astype(float)
growthrateData["yearlast"] = growthrateData["yearlast"].astype(float)
growthrateData["bingefirst"] = growthrateData["bingefirst"].astype(float)
growthrateData["bingelast"] = growthrateData["bingelast"].astype(float)

growthrateData["Growth_rate"] = (growthrateData.bingelast - growthrateData.bingefirst)/ growthrateData.bingefirst
growthrateData["Annual_Growth_rate"] = growthrateData.Growth_rate/(growthrateData.yearlast - growthrateData.yearfirst)

growthrateData_top5 = growthrateData.sort_values("Annual_Growth_rate", ascending = False)
growthrateData_top5 = growthrateData_top5.iloc[:5]


growthrateData_top5
```
